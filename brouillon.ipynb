```python
# anonymize colt data
import pandas as pd

df = pd.read_csv('./mart/private/colt.csv')
df.head(2)

leaders = df['leader'].dropna().unique()

leaders_ids_df = pd.DataFrame([{
    'leader_id': idx,
    'leader_name': leader
} for idx, leader in enumerate(leaders)])
leaders_ids_df.to_csv('./mart/private/leaders_ids.csv', index=False)


colt = pd.read_csv('./mart/private/colt.csv')
colt['leader'] = colt['leader'].map(leaders_ids_df.set_index('leader_name')['leader_id'])
colt.to_csv('./mart/private/colt.csv', index=False)

```

```python
# anonymize colt data
import pandas as pd

df = pd.read_csv('./mart/private/colt.csv')

# Group by tracks and sections to get unique types
selected_columns = ['tracks', 'section_from_id', 'section_to_id','period_type','period']
types = df[df['status'].isin(['Y', 'B'])].groupby(selected_columns).agg({
    'cou_id': lambda x: list(x)
}).reset_index()

# Add type numbers
types['type_num'] = [f"Type {i+1}" for i in range(len(types))]

# Print results
for _, row in types.iterrows():
    print(f"\n{row['type_num']}")
    for col in selected_columns: print(f"{col}: {row[col]} - ",end="")
    print('')
    print(f"cou_ids = {'; '.join(map(str, row['cou_id']))}")

print(f"Il y a {len(types)} types diffrents de coupures")
```
